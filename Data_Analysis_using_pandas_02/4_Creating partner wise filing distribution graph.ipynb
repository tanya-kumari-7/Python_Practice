```python
partner_wise_filing_count = df.groupby("sp_name")['emitra_id'].count().reset_index(name='filing_count')
partner_wise_filing_count
```

```python
fig = px.bar(
    partner_wise_filing_count,
    x = 'sp_name',
    y = 'filing_count',
    title = 'partner wise filing',
    text_auto = True ,
    color_continuous_scale='blues'
    )
fig.update_traces(
    textposition="outside",
    marker_line_color = 'black',
    marker_line_width=1. ,
) 
fig.update_layout(yaxis=dict(range=[0, partner_wise_filing_count["filing_count"].max() * 1.2]))
fig.show()
```
